# üìä Introduction to Data Visualization with Matplotlib

Data visualization is a powerful technique for representing data in a graphical format. Matplotlib is a popular data visualization library in Python that provides a wide range of plotting functions and customization options.

In this lecture, we will explore the basics of creating simple data visualizations using Matplotlib. We'll cover line plots, scatter plots, bar charts, and pie charts. By the end of this lecture, you will have a solid foundation to create your own visualizations and derive insights from data.


## üìà Line Plots

Line plots are useful for visualizing the trend or relationship between two variables. Let's start by creating a simple line plot using randomly generated data.

```python
import matplotlib.pyplot as plt

# Generate data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a line plot
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')
plt.show()
```

## üìâ Scatter Plots

Scatter plots are useful for visualizing the relationship between two continuous variables. Let's create a scatter plot to explore the relationship between the weight and height of individuals in a dataset.

```python
# Generate data
weight = [60, 65, 70, 75, 80]
height = [160, 165, 170, 175, 180]

# Create a scatter plot
plt.scatter(weight, height)
plt.xlabel('Weight (kg)')
plt.ylabel('Height (cm)')
plt.title('Weight vs. Height')
plt.show()
```

## üìä Bar Charts

Bar charts are useful for comparing categories or groups. Let's create a bar chart to visualize the sales performance of different products.

```python
# Generate data
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [500, 800, 350, 600]

# Create a bar chart
plt.bar(products, sales)
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Sales Performance')
plt.show()
```

## üçï Pie Charts

Pie charts are useful for visualizing the proportion or percentage distribution of categories. Let's create a pie chart to represent the distribution of expenses in a budget.

```python
# Generate data
categories = ['Rent', 'Groceries', 'Transportation', 'Entertainment', 'Utilities']
expenses = [1000, 500, 300, 200, 400]

# Create a pie chart
plt.pie(expenses, labels=categories, autopct='%1.1f%%', startangle=90)
plt.title('Expense Distribution')
plt.axis('equal')
plt.show()
```

## üíæ Saving Plots

Matplotlib allows you to save plots in various file formats, such as PNG, PDF, and SVG. Let's see how to save a line plot as a PNG file.

```python
# Generate data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a line plot
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')

# Save the plot as a PNG file
plt.savefig('line_plot.png')
```
